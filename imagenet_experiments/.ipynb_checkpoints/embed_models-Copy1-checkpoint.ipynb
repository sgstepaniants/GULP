```python
pretrained = True
folder = "distances/untrained"
if pretrained:
    folder = "distances/pretrained"
```

```python
def symmetrize(A):
    n = A.shape[0]
    B = A.copy()
    B[np.tril_indices(n)] = B.T[np.tril_indices(n)]
    return B
```

```python
filenames = os.listdir(folder)
distnames = []
for filename in filenames:
    if filename.endswith("npy"):
        distnames.append(filename[:-4])
distnames = np.sort(distnames)
```

```python
model = torch.load(f"models/vit_l_16.pth")
normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
transform = transforms.Compose([transforms.Resize(256), transforms.CenterCrop(224), transforms.ToTensor(), normalize])

train_dataset = datasets.ImageFolder("/home/gridsan/groups/datasets/ImageNet/train/", transform=transform)
d = model(train_dataset[0][0][None, :, :, :]).shape[1]
print(f"d = {d}")

g_cpu = torch.Generator()
g_cpu.manual_seed(1234)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=100, shuffle=True, generator=g_cpu, num_workers=0)

rep = np.zeros((n, d))
i = 0
for batch_data, batch_labels in train_loader:
    if i >= n:
        break
    nb = min(batch_data.shape[0], n-i)
    with torch.no_grad():
        rep[i:i+nb, :] = model(batch_data)[0:nb, :]
    i += nb
    print(f"{i} / {n}", flush=True)
print(flush=True)
```

```python
for distname in distnames:
    D = np.load(f'{folder}/{distname}.npy')
    D = np.reshape(D, [total_models, total_models])
    D = symmetrize(D)
    np.fill_diagonal(D, 0)
    
    Z = linkage(pdist(D), 'ward')
    #dn = dendrogram(Z)
    #plt.show()
    sorted_inds = leaves_list(Z)
    
    fig, ax = plt.subplots()
    im = ax.pcolormesh(D[sorted_inds][:, sorted_inds])
    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    fig.colorbar(im, cax=cax, orientation='vertical')
    ax.set_title(distname)
    ax.set_yticks(np.arange(total_models) + 0.5)
    ax.set_yticklabels(model_names[sorted_inds])
    ax.set_ylabel("Networks")
    ax.set_xticks(np.arange(total_models) + 0.5)
    ax.set_xticklabels(model_names[sorted_inds], rotation=90)
    ax.set_xlabel("Networks")
    plt.show()
```

```python
distname = "predictor_dist_1e-06"

D = np.load(f'{folder}/{distname}.npy')
D = np.reshape(D, [total_models, total_models])
D = symmetrize(D)
np.fill_diagonal(D, 0)

Z = linkage(pdist(D), 'ward')
sorted_inds = leaves_list(Z)
```

```python
embed_type = "UMAP"

if embed_type == "TSNE":
    X_embedded = TSNE(n_components=2, perplexity=20.0, init="random", metric="precomputed").fit_transform(D)
elif embed_type == "UMAP":
    X_embedded = umap.UMAP(n_components=2, n_neighbors=20, min_dist=0.1).fit_transform(D)
elif embed_type == "MDS":
    X_embedded = MDS(n_components=2, dissimilarity="precomputed").fit_transform(D)
```

```python
fig, ax = plt.subplots()
# plot distance matrix ordered by hierarchical clustering
im = ax.pcolormesh(D[sorted_inds][:, sorted_inds])
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im, cax=cax, orientation='vertical')
ax.set_title(distname)
ax.set_yticks(np.arange(total_models) + 0.5)
ax.set_yticklabels(model_names[sorted_inds])
ax.set_ylabel("Networks")
ax.set_xticks(np.arange(total_models) + 0.5)
ax.set_xticklabels(model_names[sorted_inds], rotation=90)
ax.set_xlabel("Networks")
plt.show()

# plot hierarhical clustering
fig, ax = plt.subplots()
dendrogram(Z, labels=model_names, ax=ax)
ax.set_title("Dendogram")
plt.show()

# plot network embedding
plt.figure(dpi=1000)
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], s=2)
for i in range(total_models):
    plt.text(X_embedded[i, 0], X_embedded[i, 1], model_names[i], fontsize=4, horizontalalignment="center", verticalalignment="bottom")
plt.title(f"{embed_type} embedding")
plt.show()
```
